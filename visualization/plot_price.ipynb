```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
input_file = "../data_processed/BTC_OKX_1000ms.csv"
```

```python
ts = pd.read_csv(input_file)

ts["datetime"] = pd.to_datetime(ts["timestamp"], unit="ms")
```

```python
# Построение графика
plt.figure(figsize=(12, 4))
plt.plot(ts["datetime"], ts["price"], label="Price", linewidth=1.5)

plt.title("Временной ряд цены")
plt.xlabel("Время")
plt.ylabel("Цена")
plt.grid(True)
plt.tight_layout()
plt.show()
```
