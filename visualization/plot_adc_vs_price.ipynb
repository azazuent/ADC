```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plot_price_vs_adc(df, price_col="price", adc_col="adc", timestamp_col="timestamp", start=None, end=None):
    """
    Рисует график цены и соответствующего ADC на общем временном интервале.
    
    Args:
        df: DataFrame с колонками цены, adc и timestamp
        price_col: имя колонки с ценами
        adc_col: имя колонки с ADC
        timestamp_col: имя колонки времени
        start: индекс начала (или timestamp)
        end: индекс конца (или timestamp)
    """
    # Срез
    plot_df = df.copy()
    if start is not None or end is not None:
        plot_df = plot_df.loc[start:end]

    fig, axs = plt.subplots(3, 1, figsize=(14, 6), sharex=True)

    # Цена
    axs[0].plot(plot_df[timestamp_col], plot_df[price_col], label="Price")
    axs[0].set_ylabel("Price")
    axs[0].legend()
    axs[0].grid(True)

    # ADC
    axs[1].plot(plot_df[timestamp_col], plot_df[adc_col], label="ADC", color='orange')
    axs[1].set_ylabel("Aggregated Directional Change")
    axs[1].legend()
    axs[1].grid(True)
    axs[1].axhline(0, color="red", linestyle="--")

    plt.xlabel("Timestamp")
    plt.tight_layout()
    plt.show()
```

```python
price_df = pd.read_csv("../data/prices/BTC_1000ms.csv")
adc_df = pd.read_csv("../data/adc/BTC_1000ms_w100.csv")
df = pd.concat([price_df, adc_df], axis = 1)
plot_price_vs_adc(df, start=8000, end=20000)
```
